```python
data = []
choice = 'random'

def line():
    print("Welcome to Daffodil International University Course Management System")
    print("======================================================================")

def show_menu():
    line()
    print("Menu")
    line()
    print("1. Enroll a Course")
    print("2. Unenroll a Course")
    print("3. View Courses")
    print("4. Exit")
    line()
    choice = input("Enter your choice: ")
    return choice

while choice != '4':
    choice = show_menu()
    if choice == '1':
        course = input("Which course do you want to get enrolled?: ")
        data.append(course)
        print("Course", course, "is enrolled successfully")

    elif choice == '2':
        course = input("Which course do you want to get unenrolled?: ")
        if course in data:
            data.remove(course)
            print("Course", course, "is unenrolled successfully")
        else:
            print("Colud not find the Course", course)
    
    elif choice == '3':
        print("Your courses are:")
        for course in data:
            print(course)
    
    elif choice == '4':
        print("Goodbye!")

    else:
        print("Please enter one of 1, 2, 3, 4")
```
